This notebook is a quick look at the LCS features in OpenDrift. GIOPS lat/lon data is used.

```python
import datetime
import os

from opendrift.readers import reader_netCDF_CF_generic
from opendrift.models.oceandrift import OceanDrift

%matplotlib inline
```

```python
data_dir='/home/soontiensn/data/MEOPAR-TREX/giops/'
```

```python
reader_U = reader_netCDF_CF_generic.Reader(os.path.join(data_dir, '*vozocrtx*.nc'))
reader_V = reader_netCDF_CF_generic.Reader(os.path.join(data_dir, '*vomecrty*.nc'))
```

# Northwest Atlantic - no land in domain

```python
o1 = OceanDrift(loglevel=20)
o1.add_reader([reader_U, reader_V])

lcs1 = o1.calculate_ftle(time=reader_U.start_time + datetime.timedelta(hours=21),
                         time_step=datetime.timedelta(minutes=30),
                         duration=datetime.timedelta(hours=12),
                         delta=0.25,
                         domain=[310,330, 40, 58],
                         RLCS=False)
```

```python
o1.plot(lcs=lcs1, colorbar=True, show_particles=False)
```

# Northwest Atlantic - land in domain

```python
o2 = OceanDrift(loglevel=20)
o2.add_reader([reader_U, reader_V])
o2.set_config('seed:ocean_only', False)
lcs2 = o2.calculate_ftle(time=reader_U.start_time + datetime.timedelta(hours=21),
                         time_step=datetime.timedelta(minutes=30),
                         duration=datetime.timedelta(hours=12),
                         delta=0.25,
                         domain=[301,330, 40, 58],
                         RLCS=False)
```

```python
o2.plot(lcs=lcs2, colorbar=True, show_particles=False, vmax=lcs1['ALCS'].max(), vmin= lcs1['ALCS'].min())
```
